```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
import re
import os
```

```python
def get_track_list(artist,album):

    url = 'https://genius.com/albums/' + re.sub(r' ', '-', artist) + '/' + re.sub(r' ', '-', album)
    page = requests.get(url)
    html = BeautifulSoup(page.text, 'html.parser')

    song_meta = html.find_all('div', class_='chart_row-content')
    
    lst = []
    
    for i in range(len(song_meta)):
        var = song_meta[i].get_text().replace('\n', ' ').replace('Lyrics', ' ').strip() \
        .encode('ascii', 'ignore')
        var = str(var).lstrip("b")
        var = re.sub('[\"]', '', var)
        var = re.sub('[\']', '', var)
        var = re.sub(' \(Ft.*', '', var)
        lst.append(var)
    
    return(lst)
```

```python
get_track_list('kanye west','the life of pablo')
```

```python
# https://medium.com/analytics-vidhya/how-to-scrape-song-lyrics-a-gentle-python-tutorial-5b1d4ab351d2

def get_lyrics(artist, album, song):
    url = 'https://genius.com/' + re.sub(r' ', '-', artist) + '-' + re.sub(r' ', '-', song) + '-lyrics'
    page = requests.get(url)
    html = BeautifulSoup(page.text, 'html.parser')
    lyrics = html.find('div', class_='lyrics').get_text()
    #remove identifiers like chorus, verse, etc
    lyrics = re.sub(r'[\(\[].*?[\)\]]', '', lyrics)
    #remove empty lines
    lyrics = os.linesep.join([s for s in lyrics.splitlines() if s])         
    lyrics = lyrics.encode('ascii', 'ignore').decode("utf-8")
    lyrics = lyrics.replace('\n', ' ')
    lyrics = lyrics.lower()
    lyrics = lyrics.split(' ')
    return(lyrics)
```

```python
get_lyrics('taylor swift','folklore','cardigan')
```
